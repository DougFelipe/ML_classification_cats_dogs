```python
## Carregar as Libs necessarias
import sys

# Adicione o caminho do diretório onde o Keras está instalado
sys.path.append('c:\\msys64\\mingw64\\lib\\python3.10\\site-packages')

## Carregar as Libs necessarias
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16, preprocess_input
import numpy as np
```

```python
## Carregar o modelo de CNN pre-treinado
model = VGG16(weights='imagenet', include_top=False, pooling='max')
```

```python
## Carregar e processar imagem
from skimage import io
from skimage.transform import resize

imagem = io.imread("american_bulldog_24.jpg")
print(imagem.shape)

imagem_reduzida = resize(imagem, (128,128))
print(imagem_reduzida.shape)
```

```python
## Colocando a imagem no formato correto
x = image.img_to_array(imagem_reduzida)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
## Extração de características
features = model.predict(x)
```

```python
## Tornar as caracteristicas em forma de vetor
features_flatten = features.flatten()
print(features_flatten.shape)
```

```python
## Carregar e processar imagem
import os
import numpy as np
import pandas as pd
from skimage import io
from skimage.transform import resize

# Listas para armazenar as características e os nomes dos arquivos
feature_list = []
filename_list = []

# Define o diretório
directory_path = "D:\\Tecnologia da Informação (TI)\\P4 2023.2\\Aprendizado de Máquina\\ProjetoCatsDogs\\dataset"

# Itera sobre cada arquivo no diretório
for filename in os.listdir(directory_path):
    # Verifica se o arquivo é uma imagem (considerando extensões comuns de imagem)
    if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
        file_path = os.path.join(directory_path, filename)
        
        # Carregar e processar imagem
        imagem = io.imread(file_path)
        imagem_reduzida = resize(imagem, (128, 128))
        
        # Colocando a imagem no formato correto
        x = image.img_to_array(imagem_reduzida)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        
        # Extração de características
        features = model.predict(x)
        
        # Adiciona características à lista, remodelando-as para serem unidimensionais
        feature_list.append(features.reshape(-1))
        
        # Adiciona o nome do arquivo à lista de nomes de arquivos
        filename_list.append(filename)

# Convertendo a lista de características em um DataFrame
df = pd.DataFrame(data=feature_list)

# Adicionando a coluna de nomes de arquivos ao início do DataFrame
df.insert(0, 'filename', filename_list)



print("Processamento concluído e características, junto com os nomes dos arquivos!")
```

```python
teste = df
# Renomeia colunas que possuem um valor numérico

teste.rename(columns={col: "feature_" + str(col) for col in teste.columns if str(col).isnumeric()}, inplace=True)
teste.rename(columns={"filename": "classe"}, inplace=True)

#modificar o nome das raças para cats e dogs
teste['classe'] = np.where(teste['classe'].str.contains('Bengal|Sphynx'), 'cats', 'dogs')

# Salvando o DataFrame como CSV
teste.to_csv("CNN_features_extracted_AVG.csv", index=False)
```
